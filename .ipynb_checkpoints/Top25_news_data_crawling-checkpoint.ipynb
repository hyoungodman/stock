```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawler(company_code, company_name):
    result = pd.DataFrame(columns=["날짜", "종목명", "기사제목"])  # "종목명" 컬럼 추가
    url = f"https://finance.naver.com/item/news_news.nhn?code={str(company_code)}&page=1"
    page = 1
    prev_data = None  # 이전에 크롤링한 데이터 저장
    
    while True:
        source_code = requests.get(url).text
        html = BeautifulSoup(source_code, "lxml")

        titles = html.select('.title')
        title_result = [re.sub('\n', '', title.get_text()) for title in titles]

        dates = html.select('.date')
        date_result = [date.get_text() for date in dates]

        df_result = pd.DataFrame({"날짜": date_result, "종목명": company_name, "기사제목": title_result})  # "종목명" 컬럼 값 설정
        
        # 중복 데이터 확인
        if prev_data is not None and df_result.equals(prev_data):
            print("마지막 페이지 도달")
            break
        
        result = pd.concat([result, df_result], ignore_index=True)
        prev_data = df_result
        
        page += 1
        url = f"https://finance.naver.com/item/news_news.nhn?code={str(company_code)}&page={page}"
        
        # 한 페이지가 완료될 때마다 결과 출력
        print(f"Page {page - 1} 완료")
    
    return result

crawled_news = {}

for i, code in enumerate(top25_stock['Code']):
    company_name = top25_stock.iloc[i]['Name']
    crawled_news[company_name] = crawler(code, company_name)  # 종목명을 함수에 전달
    print(f"{company_name} 크롤링 완료")
    print("---------------------")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawler(company_code):
    result = pd.DataFrame(columns=["날짜", "기사제목"])
    url = f"https://finance.naver.com/item/news_news.nhn?code={str(company_code)}&page=1"
    page = 1
    prev_data = None  # 이전에 크롤링한 데이터 저장
    
    while True:
        source_code = requests.get(url).text
        html = BeautifulSoup(source_code, "lxml")

        titles = html.select('.title')
        title_result = [re.sub('\n', '', title.get_text()) for title in titles]

        dates = html.select('.date')
        date_result = [date.get_text() for date in dates]

        df_result = pd.DataFrame({"날짜": date_result, "기사제목": title_result})
        
        # 중복 데이터 확인
        if prev_data is not None and df_result.equals(prev_data):
            print("마지막 페이지 도달")
            break
        
        result = pd.concat([result, df_result], ignore_index=True)
        prev_data = df_result
        
        page += 1
        url = f"https://finance.naver.com/item/news_news.nhn?code={str(company_code)}&page={page}"
        
        # 한 페이지가 완료될 때마다 결과 출력
        print(f"Page {page - 1} 완료")
    
    return result

crawled_news = {}

for i, code in enumerate(top25_stock['Code']):
    company_name = top25_stock.iloc[i]['Name']
    crawled_news[company_name] = crawler(code)
    print(f"{company_name} 크롤링 완료")
    print("---------------------")
```

```python
import os

# 저장할 디렉토리 경로
save_directory = "C:\\Crawling2023\\stock\\Top25_news_data"

def save_to_csv(data, directory, filename):
    os.makedirs(directory, exist_ok=True)
    filepath = os.path.join(directory, filename)
    data.to_csv(filepath, index=False, encoding='utf-8-sig')  # UTF-8 인코딩으로 저장
    print(f"{filename} 저장 완료")

# 데이터 저장
for company_name, news_data in crawled_news.items():
    csv_filename = f"{company_name}_news.csv"
    save_to_csv(news_data, save_directory, csv_filename)

print("데이터 저장이 완료되었습니다.")
```

```python
# 저장할 디렉토리 경로
save_directory = 'C:/Crawling2023/stock/Top25_news_data'
os.makedirs(save_directory, exist_ok=True)

# 전체 데이터를 하나의 DataFrame으로 병합
merged_data = pd.concat(crawled_news.values(), ignore_index=True)

# CSV 파일로 저장
save_path = os.path.join(save_directory, 'crawled_news.csv')
merged_data.to_csv(save_path, index=False,  encoding='utf-8-sig')

print("데이터 저장 완료:", save_path)
```

# 주석

네, 중복 데이터가 없고 뉴스가 없는 경우에도 크롤링은 알아서 종료됩니다. 이는 다음과 같은 로직으로 이루어지기 때문입니다:

1. 페이지를 크롤링하여 뉴스 데이터를 가져옵니다.
2. 가져온 뉴스 데이터가 비어있는지 확인합니다.
3. 만약 뉴스 데이터가 비어있다면, "마지막 페이지 도달" 메시지를 출력하고 while 루프를 종료합니다.
4. 뉴스 데이터가 비어있지 않다면, 가져온 데이터를 결과에 추가하고 다음 페이지로 이동합니다.
5. 다음 페이지로 이동하여 크롤링을 반복합니다.

따라서, 뉴스 데이터가 비어있어도 이전 페이지에서 중복 데이터가 발생하지 않는 이상, while 루프를 한 번 더 실행하고 "마지막 페이지 도달" 메시지를 출력하여 크롤링을 종료합니다.
