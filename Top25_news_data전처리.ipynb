```python
import pandas as pd

# crawled_news_with_company.csv 파일 불러오기
df = pd.read_csv('crawled_news_with_company.csv')

# '날짜' 열을 날짜 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'])

# 2022년 6월 13일 데이터가 있는 종목들만 선택
selected_stocks = df[df['날짜'].dt.year == 2022][df['날짜'].dt.month == 6]['종목명'].unique()

# 선택된 종목들에 해당하는 데이터만 추출
df_selected = df[df['종목명'].isin(selected_stocks)]

# 제거된 종목들 출력
removed_stocks = list(set(df['종목명'].unique()) - set(selected_stocks))
print("제거된 종목들:")
for stock in removed_stocks:
    print(stock)

# 데이터를 CSV 파일로 저장
df_selected.to_csv('filtered_news.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

# CSV 파일 로드
df = pd.read_csv('filtered_news.csv')

# 종목별 뉴스 개수 계산
news_count_by_company = df['종목명'].value_counts()

# 전체 크롤링된 뉴스 수 계산
total_news_count = len(df)

# 종목별 뉴스 개수 및 전체 뉴스 수 출력
for company_name, news_count in news_count_by_company.items():
    print(f"{company_name}: {news_count}개의 뉴스 크롤링됨")

print(f"전체 크롤링된 뉴스 수: {total_news_count}개")
```
